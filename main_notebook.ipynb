```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = pd.read_excel('docs/prods.xlsx',names=['prod', 'cant'])
base_url = 'https://listado.mercadolibre.com.co/'
```

```python
c= 0
for i in data['prod']:
  print(i[:50].replace(" ", "-").replace("\n", "-").lower(), '-------', c)
  c+=1
```

```python

def searcher(url, name):
    print(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    content = soup.find_all('li', class_='ui-search-layout__item')

    if not content:
        print("\nTermino el scraping.")
        return
    
    res = []

    for post in content[:5]:
        title = post.find('h2').text
        price = post.find('span', class_='andes-money-amount__fraction').text
        post_link = post.find("a")["href"]
        try:
            img_link = post.find("img")["data-src"]
        except:
            img_link = post.find("img")["src"]
        
        res.append({
            "name": name,
            "title": title,
            "price": price,
            "post_link": post_link,
            "image_link": img_link
        })

        
    return res

prueba = searcher(f"{base_url}juegos-llaves-bristol-1/16-3/8-lasrgas-12-pcs-medi", 'nombew')
print(prueba)
print(len(prueba))
            
            
            
```

```python
def path_to_image_html(path):
    return '<img src="'+ path + '" width="300">'

pd.set_option('display.max_colwidth', None)
```

```python
result = pd.DataFrame(columns=['title','price','post_link','image_link'])
for i in data['prod']:
  name = i[:90].replace(' ', '-').replace('\n', '-').lower()
  new = pd.DataFrame(searcher(f"{base_url}{name}", name), )
  result = pd.concat([new,result.loc[:]])
result.reset_index()
```

```python
format_dict = {}
format_dict['image_link'] = path_to_image_html
```

```python
result.to_html('test_html.html', escape=False, formatters=format_dict, classes='table { border-collapse: collapse; width: 50%; margin: 0 auto; } th, td { padding: 8px; text-align: left; border-bottom: 1px solid #ddd; } th { background-color: #f2f2f2; } tr:nth-child(even) { background-color: #f2f2f2; }')
```
